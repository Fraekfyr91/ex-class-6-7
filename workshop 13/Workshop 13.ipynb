# Cournot Model Example
- Two companies
- Linear demand
$$ p_i(q_i, q_j) = X - a q_i - bq_j,$$
where $X$ is the price when the quantity is zero, $a$ and $b$ is the marginal price effect for goods produced by company $i$ and $j$ respectively.
- Company $i$'s profit function is then given by:
$$ \pi(q_i, q_j) = (X-aq_i - bq_j)q_i -c_iq_i, $$
such that the profit $\pi$ is equal to the price times quantity minus the production costs.

- The companies best responce is the quantity that maximize profits.
- This gives a vector of fixed point, $q$, that provides the equlibrium quantities for each companies best responces.
- The Nash equilibrium is then the fixed points that solves the best responce for both companies.
